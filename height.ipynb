```python
class Records:
    details={}
    def __init__(self,name,cls,section,height):
        self.name=name
        self.height=height
        self.cls=cls
        self.section=section
        Records.details[self.name]=[self.cls,self.section,self.height]
        
    def searching_name(name):
        if name in Records.details:
            print(f"{name} : {Records.details[name][0]}  {Records.details[name][1]}  {Records.details[name][2]}")
        else:
            print("Name not in records!")
                  
    def display_data():
        for key in Records.details:
            print(f"{key} : {Records.details[key][0]}{Records.details[key][1]}  {Records.details[key][2]}")

    def modify_data(name,cls,section,height):
        Records.details[name]=[cls,section,height]

    def delete(name):
        if name in Records.details:
            Records.details.pop(name)
            print("Record deleted!!")
        else:
            print("Enter a valid name!!!")
            
    
    
        
while True:
    print("\n-----------Height Program----------\n")
    print("1. Record height")
    print("2. Display all records")
    print("3. Search by name")
    print("4. Update Data")
    print("5. Delete a record")
    print("6. Exit")
    print("\n------------------------------------\n")
    
    choice=input("Enter your choice: ")
                  
    if choice=='1':
        name=input("Enter NAME: ")
        cls=input("Enter class: ")
        section=input("Enter Section: ")
        height=float(input("Enter height in centimeter: "))
        Records(name,cls,section,height)
        continue
                  
    elif choice=='2':
        Records.display_data()
        continue
                  
    elif choice=='3':
        name=input("Enter the name: ")
        Records.searching_name(name)
        continue
    
    elif choice == '4':
        name=input("Enter NAME: ")
        cls=input("Enter class: ")
        section=input("Enter Section: ")
        height=float(input("Enter height in centimeter: "))
        Records.modify_data(name,cls,section,height)

    elif choice=='5':
        name=input("Enter the name: ")
        Records.delete(name)
        continue
                  
    elif choice=='6':
        break
    
    else:
        continue
```
