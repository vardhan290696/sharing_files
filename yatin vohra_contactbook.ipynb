```python
print("\n\n---------Contact Book----------\n\n")
phone_book={}

def save_contact(name,number):
    flag_copy = 1
    if name in phone_book:
        print("---------------------------\n")
        print("\nError\nName already exits")
        print("---------------------------\n")
        flag_copy = 0
    if number in phone_book.values():
        print("\nError\nContact already exits\n")
        print("---------------------------\n")
        flag_copy = 0
    if flag_copy == 1:
        phone_book[name]=number
        print("\n--------------------")
        print("name:",name,"--","number",number)
        print("Contact Saved Successfully")
        print("--------------------\n")

def search_contact(contact_name):
    print("\n------------------")
    if contact_name in phone_book.keys():
        print("1.",contact_name,phone_book[contact_name])
    else:
        print("No contact Found with name",contact_name)
    print("------------------\n")

def display_contact():
    serial_num=1
    sort_names=list(phone_book.keys())
    sort_names.sort()
    for count_names in phone_book:
        print(serial_num,sort_names[serial_num-1],phone_book[sort_names[serial_num-1]])
        serial_num=serial_num+1

def display_pop(value_given,opt_given):
    if opt_given == 2:
        for irt in phone_book:
            if phone_book[irt]==int(value_given):
                value_given=irt
                break
    phone_book.pop(value_given)
    
def main_operation(opt_choosen):
    if opt_choosen == 1:
        try:
            contact_name=input("Enter the Name: ")
            search_contact(contact_name)
        except ValueError:
            print("\nenter a valid value")
            print("-------------------\n")

    if opt_choosen == 2:
        print("\n------Contact Details------")
        serial_counter=1
        display_contact()
        print("\n--------------------\n\n")

    if opt_choosen  == 3:
        try:
            contact_name=input("Enter the Name: ")
            contact_num=int(input("Enter the number:"))
            save_contact(contact_name,contact_num)

        except ValueError:
            print("\nenter a valid value")
            print("-------------------\n")
    
    if opt_choosen == 4:
        print("Delete your contact:")
        print("1. By name ")
        print("2. By number ")
        op_list=[1,2]
        try:
            opt_given=int(input("Enter the input: "))
            if opt_given in op_list:
                val_given=input("Enter the Name/Number: ")
                display_pop(val_given,opt_given)
                print("\n-------------------")
                print("Contact deleted Successfully")
                print("-------------------\n")
            
            else:
                print("\n---------------------")
                print("Try again with appropriate Value")
                print("-------------------\n")

        except ValueError:
            print("\nenter a valid value")
            print("-------------------\n")



while True:
    print("1.Search Contact")
    print("2.Display Contacts")
    print("3.Save Contact")
    print("4.Delete Contact")
    print("5.Exit\n")
    option_list=[1,2,3,4,5]
    try:
        opt_choosen=int(input())
        if opt_choosen in option_list:
            main_operation(opt_choosen)
        if opt_choosen  == 5:
            break
        else:
            print("\n-------------------")    
            print("Choose the option above")
            print("-------------------\n")
    except ValueError:
        print("\n-------------------")
        print("enter a valid value")
        print("-------------------\n")
        continue
```
