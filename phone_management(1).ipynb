```python
"""This python file will manage all the records of the store"""

dic_objname = {}


class Phone:
    phone_dict = {}

    def __init__(self, ph_name, ph_company, ph_price, ph_stock):
        self.ph_name = ph_name
        self.ph_price = ph_price
        self.ph_company = ph_company
        self.ph_stock = ph_stock
        Phone.phone_dict[ph_name] = [ph_price, ph_stock, ph_company]

    def display_stock():
        print("\n\n\t\t---------Inventory------------")
        print("\nPhone Models\t\tPhone Price\tPhone Stock")
        for phone_details in Phone.phone_dict:
            if Phone.phone_dict[phone_details][1] > 0:
                print(phone_details+"\t\t"+str(Phone.phone_dict[phone_details][0])+"\t\t"+str(
                    Phone.phone_dict[phone_details][1]))
        print("\n\t\t-----------------------------\n")

    def sold_phone(self, peice_sold):
        if self.ph_stock >= peice_sold:
            self.ph_stock = self.ph_stock-peice_sold
            return 1
        else:
            return 0


class Sales(Phone):
    turn_over = 0


def display_turnover():
    print("\n\t\t---------Turn Over---------\n")
    print("\t\t\tRs.", Sales.turn_over)
    print("\n\t\t---------------------------\n")


def display_stock():
    Phone.display_stock()


def search_by_brand(brand_name):
    flag_check = 0
    print(f"\n\n\t\t------Inventory[{brand_name}]---------")
    print("\nPhone Models\t\tPhone Price\tPhone Stock")
    for phone_details in Phone.phone_dict:
        if Phone.phone_dict[phone_details][2] == brand_name:
            if Phone.phone_dict[phone_details][1] > 0:
                flag_check = 1
                print(phone_details+"\t\t"+str(Phone.phone_dict[phone_details][0])+"\t\t"+str(
                    Phone.phone_dict[phone_details][1]))
    print("\n\t\t-----------------------------\n")


def update_stock(obj_name, stock_val, option_stock="add"):
    flag_check = 0
    if stock_val > 0:
        if option_stock == "add":
            obj_name.ph_stock = obj_name.ph_stock + stock_val
        elif option_stock == "sub":
            obj_name.ph_stock = obj_name.ph_stock - stock_val
        Phone.phone_dict[obj_name.ph_name] = [
            obj_name.ph_price, obj_name.ph_stock, obj_name.ph_company]
        flag_check = 1
    else:
        pass
    inventory_update(obj_name, flag_check)


def inventory_update(obj_name, flag_val):
    print("\n---------------------------")
    if flag_val == 1:
        print("Inventory Updated Successfully", obj_name.ph_name)
    else:
        print("Enter the Correct Value")
    print("---------------------------\n")


def add_phone(obj_name):
    dic_objname[obj_name.ph_name] = obj_name


def remove_item(obj_name):
    Phone.phone_dict.pop(obj_name.ph_name)
    inventory_update(obj_name,1)


def sold_phone(obj_name, peice_sold=1):
    flag_val = obj_name.sold_phone(peice_sold)
    Phone.phone_dict[obj_name.ph_name] = [
        obj_name.ph_price, obj_name.ph_stock, obj_name.ph_company]
    Sales.turn_over = Sales.turn_over + obj_name.ph_price
    inventory_update(obj_name, flag_val)


iphone_13 = Phone("iphone 13", "apple", 13900, 9)
add_phone(iphone_13)
iphone_12 = Phone("iphone 12", "apple", 99900, 2)
add_phone(iphone_12)
iphone_11 = Phone("iphone 11", "apple", 5900, 4)
add_phone(iphone_11)
samsung_s10 = Phone("Samsung s10", "samsung", 58000, 58)
add_phone(samsung_s10)
samsung_note_10 = Phone("Samsung Note 10", "samsung", 10900, 8)
add_phone(samsung_note_10)
vivo_v9 = Phone("Vivo v9 Pro", "vivo", 9000, 2)
add_phone(vivo_v9)
vivo_s10 = Phone("Vivo s10", "vivo", 13900, 1)
add_phone(vivo_s10)
oppo_reno3 = Phone("oppo reno 3", "oppo", 49000, 1)
add_phone(oppo_reno3)

# display_stock()
# sold_phone(oppo_reno3)
# sold_phone(samsung_s10 ,50)   #for making the peice sold
# sold_phone(samsung_note_10)  #stock is optional
# display_stock()       #for displaying the inventory
# display_turnover()    #for display the turn over
# search_by_brand("vivo")  # for display inventory by brand
# for updating the stock value, giv argument subtract to stract the stock
# update_stock(samsung_s10, 10,"sub")

print("\n\n\t\t--------Yatin Mobile Store----------")

while True:
    print("1.Display inventory")
    print("2.Sold Phone")
    print("3.Total Turn over")
    print("4.Inventory Search by Brand")
    print("5.Update stocks")
    print("6.Remove Item from inventory")
    print("7.Exit")
    opt_list = [1, 2, 3, 4, 5, 6, 7]
    try:
        opt_choosen = int(input("Option:"))
        if opt_choosen not in opt_list:
            print("\n-----------------------------")
            print("\nChoose the option above")
            print("-----------------------------\n")
            continue
    except:
        print("\nEnter a valid Value")
        continue
    if opt_choosen == 1:
        display_stock()

    elif opt_choosen == 2:
        print("\n---------------")
        phone_name = input("Phone Name: ")
        if phone_name not in dic_objname.keys():
            print("No phone with name", phone_name)
            print("\nTry with names")
            for ph_name in dic_objname:
                print(ph_name, end=",")
            print("----------------------")
        else:
            sold_phone(dic_objname[phone_name])

    elif opt_choosen == 3:
        display_turnover()

    elif opt_choosen == 4:
        print("\n---------------")
        phone_name = input("Phone Name: ")
        search_by_brand(phone_name)

    elif opt_choosen == 5:
        print("\n1.add stock")
        print("2.remove stock\n")
        op_choose = int(input("option:"))
        print("\n---------------")
        phone_name = input("Phone Name: ")
        print("\n---------------")
        stock_val = int(input("Stock value: "))
        if op_choose == 2:
            update_stock(dic_objname[phone_name], stock_val, "sub")
        else:
            update_stock(dic_objname[phone_name], stock_val)

    elif opt_choosen == 6:
        print("\n---------------")
        phone_name = input("Phone Name: ")
        if phone_name not in dic_objname.keys():
            print("\n-----------------")
            print(phone_name,"Not found in inventory")
            print("\n-----------------")
        remove_item(dic_objname[phone_name])

    elif opt_choosen == 7:
        exit()
```
